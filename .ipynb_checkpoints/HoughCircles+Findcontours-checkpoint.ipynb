```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

IMAGE_PATH = "olho.png"
img_color = cv2.imread(IMAGE_PATH)
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

plt.xticks([]), plt.yticks([])
plt.imshow(img_color)
```

```python
img = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
img = cv2.medianBlur(img,5)
# Razão entre imagem de entrada e domínio dos parâmetros (espaço de Hough)
dp = 1
# Distância mínima entre detecções
min_distance = 50
# Limiar superior para detecção de bordas
canny_thresh = 30
# Limiar de detecção de centros dos círculos
hough_tresh = 40
# Tamanho mínimo do raio
min_radius = 20
# Tamanho máximo do raio
maxRadius = 45

circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, dp, min_distance, param1=canny_thresh, param2=hough_tresh, minRadius=min_radius, maxRadius=maxRadius)

for circle in circles[0,:]:
    x, y, r = circle
    center = (int(x),int(y))
    # Desenha o círculo
    cv2.circle(img_color, center, int(r), (0,255,0), 2)
    # Desenha o centro do círculo
    cv2.circle(img_color, center, 2, (255,0,255), 3)
    
plt.xticks([]), plt.yticks([])
plt.imshow(img_color)
```

```python
import math

import cv2
import numpy as np

from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

IMAGE_PATH = "olho2.png"
img_color = cv2.imread(IMAGE_PATH)
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

ret, thresh_gray = cv2.threshold(gray, 220, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(thresh_gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    area = cv2.contourArea(contour)
    rect = cv2.boundingRect(contour)
    x, y, width, height = rect
    radius = 0.25 * (width + height)
    print (area)

    if area <= 200:
        cv2.circle(img_color, (int(x + radius), int(y + radius)), int(8*radius), (0,180,0), 2)

plt.xticks([]), plt.yticks([])
plt.imshow(img_color)
```
